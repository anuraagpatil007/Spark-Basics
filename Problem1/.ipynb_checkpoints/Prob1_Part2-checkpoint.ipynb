```python
# -*- coding: utf-8 -*-
import string
from pyspark import SparkConf

inputFileRdd = sc.wholeTextFiles("/user/shakespere/*")
switchedInputRdd = inputFileRdd.map(lambda (fPath,fContent) : (fContent,fPath))
contentPathRdd = switchedInputRdd\
                 .mapValues(lambda fPath : fPath.split('/')[-1])
    
#remove punctuation 
removePunct = lambda fContent : fContent not in string.punctuation

#creating Rdd with words(keys) without punctuation 
contentPathRdd1 = contentPathRdd.map(lambda (fContent, fName) : (filter(removePunct, fContent), fName) )

#converting to lower case 
def func(fname):
    fnameList = []
    for i in fname:
        if i not in fnameList:
            fnameList.append(i)   
    return fnameList

contentPathRdd2 = contentPathRdd1.map(lambda (x,y) : (x.lower(), y))
contentPathRdd3 = contentPathRdd2\
                  .flatMap(lambda(content,name):map(lambda word : (word,name), content.split()))
contentPathRdd4 = contentPathRdd3.map(lambda (word,name) : ((word,name), 1))
contentPathRdd5 = contentPathRdd4.reduceByKey(lambda cnt1, cnt2 : cnt1+cnt2)
contentPathRdd6 = contentPathRdd5.map(lambda((word,name), cnt):(word,(name,cnt)))


intdeOutputRdd = contentPathRdd6\
                 .groupByKey().mapValues(list)
finalOutput = intdeOutputRdd.map(lambda (word,fname):(word,func(fname)))

for i in finalOutput.collect():
    print i

file10.saveAsTextFile("/user/spark/output2/FinalOutputProb1_Part2.txt")

  
```
